```python
def data_process_hrs_betas(data):
    data.columns = ['variable', 'coef', 'exp(coef)', 'se(coef)', 'coef lower 95%', 'coef upper 95%', 'exp(coef) lower 95%', 'exp(coef) upper 95%', 'cmp to', 'z', 'p', '-log2(p)']
    data_hrs = data[['variable', 'exp(coef)', 'exp(coef) lower 95%', 'exp(coef) upper 95%']]
    data_hrs['exp(coef)'] = data_hrs['exp(coef)'].apply(lambda x: "{:.2f}".format(x))
    data_hrs['exp(coef) lower 95%'] = data_hrs['exp(coef) lower 95%'].apply(lambda x: "{:.2f}".format(x))
    data_hrs['exp(coef) upper 95%'] = data_hrs['exp(coef) upper 95%'].apply(lambda x: "{:.2f}".format(x))
    data_hrs['betas'] = data_hrs['exp(coef)'].astype(str) + " (" + data['exp(coef) lower 95%'].astype(str) + '--' + data['exp(coef) upper 95%'].astype(str) + ')'
    
    data_hrs = data_hrs.sort_values('variable')
    data_hrs = data_hrs[['variable', 'betas']]
    
    data_betas = data[['variable', 'coef']]
    
    return data_hrs, data_betas
```
