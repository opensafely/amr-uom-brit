```python
def data_process_hrs_betas(data_ab_crude, data_ab_type, data_ab_15_24_strat, data_ab_25_34_strat, data_ab_35_44_strat, data_ab_45_54_strat, data_ab_55_64_strat, data_ab_65_74_strat, data_ab_75_more_strat, data_ab_male_strat, data_ab_female_strat, data_ab_prepandemic_strat, data_ab_during_pandemic_strat, data_ab_post_2nd_lockdown_strat):#, ):# cph_lrti_incdt_ab_prepandemic_1):#, , data_ab_female_strat):
    data_ab_crude.columns = ['covariate', 'exp(coef)', '95% lower-bound', '95% upper-bound']
    data_ab_type.columns = ['covariate', 'exp(coef)', '95% lower-bound', '95% upper-bound']
    data_ab_15_24_strat.columns = ['covariate', 'exp(coef)', '95% lower-bound', '95% upper-bound']
    data_ab_25_34_strat.columns = ['covariate', 'exp(coef)', '95% lower-bound', '95% upper-bound']
    data_ab_35_44_strat.columns = ['covariate', 'exp(coef)', '95% lower-bound', '95% upper-bound']
    data_ab_45_54_strat.columns = ['covariate', 'exp(coef)', '95% lower-bound', '95% upper-bound']
    data_ab_55_64_strat.columns = ['covariate', 'exp(coef)', '95% lower-bound', '95% upper-bound']
    data_ab_65_74_strat.columns = ['covariate', 'exp(coef)', '95% lower-bound', '95% upper-bound']
    data_ab_75_more_strat.columns = ['covariate', 'exp(coef)', '95% lower-bound', '95% upper-bound']
    data_ab_male_strat.columns = ['covariate', 'exp(coef)', '95% lower-bound', '95% upper-bound']
    data_ab_female_strat.columns = ['covariate', 'exp(coef)', '95% lower-bound', '95% upper-bound']
    data_ab_prepandemic_strat.columns = ['covariate', 'exp(coef)', '95% lower-bound', '95% upper-bound']
    data_ab_during_pandemic_strat.columns = ['covariate', 'exp(coef)', '95% lower-bound', '95% upper-bound']
    data_ab_post_2nd_lockdown_strat.columns = ['covariate', 'exp(coef)', '95% lower-bound', '95% upper-bound']

    data_ab_crude = data_ab_crude[data_ab_crude['covariate'].str.contains('ab_')]
    data_ab_type = data_ab_type[data_ab_type['covariate'].str.contains('ab_')]
    data_ab_15_24_strat = data_ab_15_24_strat[data_ab_15_24_strat['covariate'].str.contains('ab_')]
    data_ab_25_34_strat = data_ab_25_34_strat[data_ab_25_34_strat['covariate'].str.contains('ab_')]
    data_ab_35_44_strat = data_ab_35_44_strat[data_ab_35_44_strat['covariate'].str.contains('ab_')]
    data_ab_45_54_strat = data_ab_45_54_strat[data_ab_45_54_strat['covariate'].str.contains('ab_')]
    data_ab_55_64_strat = data_ab_55_64_strat[data_ab_55_64_strat['covariate'].str.contains('ab_')]
    data_ab_65_74_strat = data_ab_65_74_strat[data_ab_65_74_strat['covariate'].str.contains('ab_')]
    data_ab_75_more_strat = data_ab_75_more_strat[data_ab_75_more_strat['covariate'].str.contains('ab_')]
    data_ab_male_strat = data_ab_male_strat[data_ab_male_strat['covariate'].str.contains('ab_')]
    data_ab_female_strat = data_ab_female_strat[data_ab_female_strat['covariate'].str.contains('ab_')]
    data_ab_prepandemic_strat = data_ab_prepandemic_strat[data_ab_prepandemic_strat['covariate'].str.contains('ab_')]
    data_ab_during_pandemic_strat = data_ab_during_pandemic_strat[data_ab_during_pandemic_strat['covariate'].str.contains('ab_')]
    data_ab_post_2nd_lockdown_strat = data_ab_post_2nd_lockdown_strat[data_ab_post_2nd_lockdown_strat['covariate'].str.contains('ab_')]

    data_ab_type['covariate'] = data_ab_type['covariate'].str.replace('Doxycycline', 'most_frequent')
    data_ab_15_24_strat['covariate'] = data_ab_15_24_strat['covariate'].astype(str) + '_15_24'
    data_ab_25_34_strat['covariate'] = data_ab_25_34_strat['covariate'].astype(str) + '_25_34'
    data_ab_35_44_strat['covariate'] = data_ab_35_44_strat['covariate'].astype(str) + '_35_44'
    data_ab_45_54_strat['covariate'] = data_ab_45_54_strat['covariate'].astype(str) + '_45_54'
    data_ab_55_64_strat['covariate'] = data_ab_55_64_strat['covariate'].astype(str) + '_55_64'
    data_ab_65_74_strat['covariate'] = data_ab_65_74_strat['covariate'].astype(str) + '_65_74'
    data_ab_75_more_strat['covariate'] = data_ab_75_more_strat['covariate'].astype(str) + '_75_more'
    data_ab_male_strat['covariate'] = data_ab_male_strat['covariate'].astype(str) + '_male'
    data_ab_female_strat['covariate'] = data_ab_female_strat['covariate'].astype(str) + '_female'
    data_ab_prepandemic_strat['covariate'] = data_ab_prepandemic_strat['covariate'].astype(str) + '_prepandemic'
    data_ab_during_pandemic_strat['covariate'] = data_ab_during_pandemic_strat['covariate'].astype(str) + '_during_pandemic'
    data_ab_post_2nd_lockdown_strat['covariate'] = data_ab_post_2nd_lockdown_strat['covariate'].astype(str) + '_post_2nd_lockdown'    

    data_hrs_abs = pd.concat([data_ab_crude, data_ab_type, data_ab_15_24_strat, data_ab_25_34_strat, data_ab_35_44_strat, data_ab_45_54_strat, data_ab_55_64_strat, data_ab_65_74_strat, data_ab_75_more_strat, data_ab_male_strat, data_ab_female_strat, data_ab_prepandemic_strat, data_ab_during_pandemic_strat, data_ab_post_2nd_lockdown_strat], axis = 0) #, , , data_ab_during_pandemic_strat, data_ab_female_strat

    data_hrs_abs['exp(coef)'] = data_hrs_abs['exp(coef)'].apply(lambda x: "{:.2f}".format(x))
    data_hrs_abs['95% lower-bound'] = data_hrs_abs['95% lower-bound'].apply(lambda x: "{:.2f}".format(x))
    data_hrs_abs['95% upper-bound'] = data_hrs_abs['95% upper-bound'].apply(lambda x: "{:.2f}".format(x))
    data_hrs_abs['betas'] = data_hrs_abs['exp(coef)'].astype(str) + " (" + data_hrs_abs['95% lower-bound'].astype(str) + '-' + data_hrs_abs['95% upper-bound'].astype(str) + ')'
    # data_hrs_abs_final = data_hrs_abs[['covariate', 'betas']]
    data_hrs_abs = data_hrs_abs.reset_index(drop=True)

    return data_hrs_abs#_final
```
