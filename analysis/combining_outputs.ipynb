```python
d = {'variable1': [output_lrti_no_abs_incdt.columns[0], output_urti_no_abs_incdt.columns[0], 
                   output_uti_no_abs_incdt.columns[0], output_sinusitis_no_abs_incdt.columns[0],
                   output_otmedia_no_abs_incdt.columns[0], output_ot_externa_no_abs_incdt.columns[0],
                   output_lrti_no_abs_prevl.columns[0], output_urti_no_abs_prevl.columns[0],
                   output_uti_no_abs_prevl.columns[0], 
                   output_sinusitis_no_abs_prevl.columns[0], output_otmedia_no_abs_incdt.columns[0],
                   output_ot_externa_no_abs_prevl.columns[0], 
                 ],
     'variable2': [output_lrti_no_abs_incdt.loc[0][0], output_urti_no_abs_incdt.loc[0][0],
                   output_uti_no_abs_incdt.loc[0][0], output_sinusitis_no_abs_incdt.loc[0][0],
                   output_otmedia_no_abs_incdt.loc[0][0], output_ot_externa_no_abs_incdt.loc[0][0],
                   output_lrti_abs_incdt.loc[0][0], output_urti_abs_incdt.loc[0][0],
                   output_uti_abs_incdt.loc[0][0], 
                   output_sinusitis_abs_incdt.loc[0][0], output_otmedia_abs_incdt.loc[0][0], 
                   output_ot_externa_abs_incdt.loc[0][0],
                 ]
    }
data = pd.DataFrame(data=d)

#keep concordance index and remove other letters, and shape dataframe similar to the table
data[['var1', 'ci1']] = data['variable1'].str.split(':', expand=True)
data[['var2', 'ci2']] = data['variable2'].str.split(':', expand=True)
data = data.drop(columns=['variable1', 'variable2', 'var2'])
data.columns = ['infection', 'ci_dev', 'ci_val']

#keep two decimal
data['ci_dev'] = data['ci_dev'].astype(float)
data['ci_dev'] = np.round(data['ci_dev'], decimals=2)
data['ci_val'] = data['ci_val'].astype(float)
data['ci_val'] = np.round(data['ci_val'], decimals=2)

# data['infection'] = data['infection'].str.split('with', expand=True)[0]
data['infection'] = data['infection'].str.replace('Concordance index for ', '')
data['infection'] = data['infection'].str.replace('with training data', '')
data['infection'] = data['infection'].str.replace('with testing data', '')

data
```
