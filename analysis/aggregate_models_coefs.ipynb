```python
def data_process_hrs(data):
    data_hrs = data
    data_hrs['exp(coef)'] = np.round(data_hrs['exp(coef)'], decimals=2)
    data_hrs['95% lower-bound'] = np.round(data_hrs['95% lower-bound'], decimals=2)
    data_hrs['95% upper-bound'] = np.round(data_hrs['95% upper-bound'], decimals=2)
    data_hrs['betas'] = data_hrs['exp(coef)'].astype(str) + " (" + data_hrs['95% lower-bound'].astype(str) + '--' + data_hrs['95% upper-bound'].astype(str) + ')'
    
    return data_hrs
```
